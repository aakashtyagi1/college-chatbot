```python
import random
import re
import string

# =========================
# KNOWLEDGE BASE
# =========================

KNOWLEDGE_BASE = [

    {
        "tag": "greeting",

        "patterns": [
            "hello",
            "hi",
            "hey",
            "good morning"
        ],

        "responses": [
            "Hello! Welcome to CollegeBot.",
            "Hi there! How can I help you?"
        ]
    },

    {
        "tag": "admission",

        "patterns": [
            "admission",
            "apply",
            "enroll"
        ],

        "responses": [
            "Admissions start every June.",
            "You can apply from the college website."
        ]
    },

    {
        "tag": "fees",

        "patterns": [
            "fee",
            "fees",
            "college fees"
        ],

        "responses": [
            "Engineering fees are Rs 80,000 per year.",
            "Scholarships are available."
        ]
    },

    {
        "tag": "courses",

        "patterns": [
            "courses",
            "programs"
        ],

        "responses": [
            "We offer B.Tech, BCA, MCA and MBA."
        ]
    },

    {
        "tag": "hostel",

        "patterns": [
            "hostel",
            "accommodation"
        ],

        "responses": [
            "Separate hostels are available for boys and girls."
        ]
    },

    {
        "tag": "placements",

        "patterns": [
            "placement",
            "job"
        ],

        "responses": [
            "Average placement package is Rs 6 LPA."
        ]
    },

    {
        "tag": "thanks",

        "patterns": [
            "thanks",
            "thank you"
        ],

        "responses": [
            "You are welcome!"
        ]
    }

]

# =========================
# PREPROCESS FUNCTION
# =========================

def preprocess(text):

    text = text.lower()

    text = text.translate(
        str.maketrans('', '', string.punctuation)
    )

    return text

# =========================
# REGEX PATTERNS
# =========================

REGEX_PATTERNS = [

    (r"\b(hello|hi|hey)\b", "greeting"),

    (r"\b(admission|apply|enroll)\b", "admission"),

    (r"\b(fee|fees|college fees)\b", "fees"),

    (r"\b(course|courses|programs)\b", "courses"),

    (r"\b(hostel|accommodation)\b", "hostel"),

    (r"\b(placement|placements|job)\b", "placements"),

    (r"\b(thanks|thank you)\b", "thanks")
]

# =========================
# REGEX MATCH FUNCTION
# =========================

def regex_match(user_input):

    cleaned_input = preprocess(user_input)

    for pattern, tag in REGEX_PATTERNS:

        if re.search(pattern, cleaned_input):

            for item in KNOWLEDGE_BASE:

                if item["tag"] == tag:

                    return random.choice(item["responses"])

    return "Sorry, I did not understand that."

# =========================
# MAIN CHATBOT LOOP
# =========================

print("CollegeBot Started!")
print("Type 'bye' to exit.")

while True:

    user_input = input("You: ")

    if user_input.lower() == "bye":

        print("Bot: Goodbye!")

        break

    response = regex_match(user_input)

    print("Bot:", response)
```
